```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
df = pd.read_csv("weatherAUS.csv")
print("Initial shape:", df.shape)
print(df.info())
print((df.isnull().sum()*100/(df.shape[0])).sort_values())

```

```python

threshold = 0.3 
df = df.loc[:, df.isnull().mean() < threshold]
print((df.isnull().sum()*100/(df.shape[0])).sort_values())
```

```python
df = df.dropna(subset=['Pressure9am','Pressure3pm'])
print((df.isnull().sum()*100/(df.shape[0])).sort_values())
```

```python
df=df.fillna(df.median(numeric_only=True))
df=df.dropna()
print((df.isnull().sum()*100/(df.shape[0])).sort_values())
```

```python
df.shape
```

```python


dirs = ['N','NNE','NE','ENE','E','ESE','SE','SSE','S','SSW','SW','WSW','W','WNW','NW','NNW']
dir_to_deg = {d: i*(360/16) for i, d in enumerate(dirs)}

for col in ['WindGustDir', 'WindDir9am', 'WindDir3pm']:
    df[col] = df[col].map(dir_to_deg)
    df[col + '_sin'] = np.sin(np.deg2rad(df[col]))
    df[col + '_cos'] = np.cos(np.deg2rad(df[col]))
    df.drop(col, axis=1, inplace=True)
df.head()
```

```python
df = pd.get_dummies(df, columns=['Location'], prefix='Loc',drop_first=True)
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.head()
```

```python

df['Date'].dt.year.value_counts().sort_index(),df.info()
```

```python
from sklearn.preprocessing import StandardScaler
exclude_cols = ['Date', 'RainToday', 'RainTomorrow']
exclude_cols += [col for col in df.columns if col.endswith(('_sin', '_cos'))]
exclude_cols += list(df.select_dtypes(include='bool').columns)

num_cols = df.select_dtypes(include=['float64', 'int32']).columns.difference(exclude_cols)

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("Scaled columns:", list(num_cols))
print(df[num_cols].describe().round(2))
```
